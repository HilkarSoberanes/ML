# Commly used activations functions for Convolutional Neural Networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def ReLu(x):
    return np.maximum(0,x)

def LeakyReLu(x, alpha=0.1):
    return np.maximum(0,x) + alpha*np.minimum(0,x)
```

```python
def plot(function):
    fig, ax = plt.subplots()
    # Move the left and bottom spines to x = 0 and y = 0, respectively.
    
    x = np.linspace(-5, 5, 100)
    
    bottom_height=(max(function(x)))/2
    
    
    ax.spines[["left"]].set_position(("axes", 0.5))
    ax.spines[["bottom"]].set_position(("axes",0.5))
    
    # Hide the top and right spines.
    ax.spines[["top", "right"]].set_visible(False)

    # Draw arrows (as black triangles: ">k"/"^k") at the end of the axes.  In each
    # case, one of the coordinates (0) is a data coordinate (i.e., y = 0 or x = 0,
    # respectively) and the other one (1) is an axes coordinate (i.e., at the very
    # right/top of the axes).  Also, disable clipping (clip_on=False) as the marker
    # actually spills out of the axes.
    ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
    ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

    # Some sample data.
    
    ax.plot(x, function(x), color='red')
    plt.xticks([])
    plt.yticks([])
    plt.xlim(-5,5)
    plt.ylim(-2,2)
    plt.xlabel(r'$z$', loc='right')
    plt.ylabel(r'$a(z)$', loc='top', rotation=0)
    plt.show()
```

```python
plot(sigmoid)
```

```python
plot(tanh)
```

```python
plot(ReLu)
```

```python
plot(LeakyReLu)
```
